### Question 1: Import Libraries and Load Image (1 Mark)

Import `numpy` as `np` and `cv2`. Download the image 'moon-pexels-frank-cone.jpg' from https://drive.google.com/drive/folders/1gieHICVDBbUKMZiSF4YRQMAJic-w50JM?usp=sharing and load it into a variable named `original_image` using OpenCV. Print the shape of the loaded image.

**Expected Output:** A tuple representing the shape of the image (height, width, channels).

```python
import numpy as np
import cv2

# Load the image (make sure the image file is in the current working directory)
original_image = cv2.imread('moon-pexels-frank-cone.jpg')

# Print the shape of the image
print(original_image.shape)
```

### Question 2: Convert to Grayscale (2 Marks)

Convert the `original_image` to grayscale using OpenCV's `cvtColor` function. Store the result in a variable named `grayscale_image`. Print the shape of the new grayscale image.

**Expected Output:** A tuple representing the shape of the grayscale image (height, width).

```python
import cv2

# Convert the original image to grayscale
grayscale_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2GRAY)

# Print the shape of the grayscale image
print(grayscale_image.shape)
```

### Question 4: Shift the Image (2 Marks)

Create a transformation matrix `M` to shift the `original_image` 50 pixels to the right and 100 pixels down. Apply this transformation using `cv2.warpAffine` and store the result in `shifted_image`. Print the shape of `shifted_image`.

**Expected Output:** The shape of the shifted image, which will be identical to the original image's shape.

```python
import numpy as np
import cv2

# Create the translation matrix
M = np.float32([
    [1, 0, 50],   # Shift 50 pixels to the right
    [0, 1, 100]   # Shift 100 pixels down
])

# Apply the translation
shifted_image = cv2.warpAffine(
    original_image,
    M,
    (original_image.shape[1], original_image.shape[0])  # (width, height)
)

# Print the shape of the shifted image
print(shifted_image.shape)
```

### Question 5: Resize the Image (1 Mark)

Resize the `original_image` to be 150 pixels wide and 100 pixels tall. Store the result in a variable named `resized_image` and print its new shape.

**Expected Output:** The tuple `(100, 150, 3)` representing the new shape (height, width, channels).

```python
import cv2

# Resize the image to width=150 and height=100
resized_image = cv2.resize(original_image, (150, 100))

# Print the new shape
print(resized_image.shape)
```

### Question 6: Rotate the Image (2 Marks)

Rotate the `original_image` by 90 degrees counter-clockwise around its center. Store the result in a variable named `rotated_image` and print its shape.

**Expected Output:** The shape of the rotated image. Note that the height and width will be swapped compared to the original.

```python
import cv2

# Get the image dimensions
(h, w) = original_image.shape[:2]

# Compute the center of the image
center = (w // 2, h // 2)

# Create the rotation matrix for 90 degrees counter-clockwise
M = cv2.getRotationMatrix2D(center, 90, 1.0)

# Apply the rotation, ensuring dimensions are swapped for the output size (width, height)
rotated_image = cv2.warpAffine(original_image, M, (h, w))

# Print the shape of the rotated image
print(rotated_image.shape)
```

### Question 7: Create an Image Transform Pipeline (2 Marks)

Import `torch` and necessary modules from `torchvision`. Define a transform pipeline named `train_transform` that resizes images to 255x255, randomly flips them horizontally, and then converts them to a tensor.

**Hint:** Use `transforms.Compose()` with a list containing `transforms.Resize()`, `transforms.RandomHorizontalFlip()`, and `transforms.ToTensor()`.

**Expected Output:** No output, but the `train_transform` object should be created successfully.

```python
import torch
from torchvision import transforms

# Define the training transform pipeline
train_transform = transforms.Compose([
    transforms.Resize((255, 255)),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor()
])
```

### Question 8: Create an ImageFolder Dataset (2 Marks)

Create an `ImageFolder` dataset named `train_dataset` from the `'Cat_Dog_data/train'` directory, applying the `train_transform` pipeline you just created. Print the total number of images found in the dataset.

**Hint:** Use `datasets.ImageFolder(data_dir, transform=your_transform)`.

**Expected Output:** A printout of the number of images in the training dataset.

```python
from torchvision import datasets

# Path to training data
data_dir = "Cat_Dog_data/train"

# Create ImageFolder dataset
train_dataset = datasets.ImageFolder(root=data_dir, transform=train_transform)

# Print number of images
print(len(train_dataset))
```

### Question 9: Create a DataLoader (4 Marks)

Create a `DataLoader` named `train_loader` from the `train_dataset`. Set the `batch_size` to 64 and `shuffle` to True. Then, retrieve one batch of images and labels from the loader and print the shape of the images tensor and the labels tensor.

**Hint:** Use `torch.utils.data.DataLoader()`. To get one batch, use `images, labels = next(iter(train_loader))`.

**Expected Output:** Two printed tuples representing the shapes of the image batch and the label batch.

```python
import torch
from torch.utils.data import DataLoader

# Create DataLoader
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)

# Get one batch
images, labels = next(iter(train_loader))

# Print shapes
print(images.shape)
print(labels.shape)
```

### Question 10: Load MNIST Data (3 Marks)

Load the MNIST dataset using `sklearn.datasets.load_digits`. Separate the dataset into features (`X`) and target labels (`y`).
Print the shape of the features and the target arrays.

**Expected Output:** The shape of `X` and `y`.

```python
from sklearn.datasets import load_digits

# Load MNIST (digits) dataset
digits = load_digits()

# Features and labels
X = digits.data
y = digits.target

# Print shapes
print(X.shape)
print(y.shape)
```

### Question 11: K-Means Clustering (5 Marks)

Import `KMeans` from `sklearn.cluster`. Initialize and fit the K-Means clustering model on the MNIST features (`X`).
Since we know there are 10 digits (0-9), set the number of clusters to 10. Assign the cluster labels to a variable `kmeans_labels`.

**Expected Output:** A successfully fitted K-Means model and the cluster labels array.

```python
from sklearn.cluster import KMeans

# Create K-Means model with 10 clusters
kmeans = KMeans(
    n_clusters=10,
    random_state=42,
    n_init=10
)

# Train the model
kmeans.fit(X)

# Get cluster labels
kmeans_labels = kmeans.labels_

# Print cluster labels
print(kmeans_labels)

# Print shape of labels array
print("Shape of cluster labels:", kmeans_labels.shape)
```

### Question 12: F1 Score Evaluation for Clustering (5 Marks)

Evaluate the clustering performance using the F1 score. Since K-Means assigns arbitrary cluster labels, you will first need to map each cluster label to the most frequent true label in that cluster.
After mapping the labels, calculate and print the macro-averaged F1 score using `sklearn.metrics.f1_score`.

**Expected Output:** The calculated F1 score of the clustering.

```python
from sklearn.metrics import f1_score
from scipy.stats import mode
import numpy as np

# Create mapping from cluster -> true label
cluster_to_label = {}

for i in range(10):
    mask = (kmeans_labels == i)

    if np.sum(mask) > 0:
        cluster_to_label[i] = mode(y[mask], keepdims=True).mode[0]

# Convert cluster labels to predicted labels
predicted_labels = np.array([
    cluster_to_label[label] for label in kmeans_labels
])

# Calculate F1 score
f1 = f1_score(y, predicted_labels, average='macro')

print("Macro F1 Score:", f1)
```
